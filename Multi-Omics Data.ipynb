```python
# Import 'PCA' from 'sklearn' library
from sklearn.decomposition import PCA

# Import 'matplotlib' library
import matplotlib.pyplot as plt

# Import 'mplot3d' library
from mpl_toolkits import mplot3d
```

```python
# Perform Principal Component Analysis on raw data
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(X)
principal_DF = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
```

```python
# Plotting principal components
%matplotlib notebook
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(principal_DF['principal component 1'], principal_DF['principal component 2'], principal_DF['principal component 3'], c=y, cmap="bwr", s=25)
 
# label the axes
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("PCA on Multi-omic Dataset")
plt.show()
```

```python
# Import 'mutual_info_classif' and 'SelectKBest' from 'sklearn' library
from sklearn.feature_selection import mutual_info_classif,SelectKBest
```

```python
# Select best 250 features from 300 features set using 'MI' method
fs = SelectKBest(score_func=mutual_info_classif, k=250)
fs.fit(X,y)
cols = fs.get_support(indices=True)
features_250_X = X.iloc[:,cols]
```

```python
# Select best 150 features from 200 features set using 'MI' method
fs = SelectKBest(score_func=mutual_info_classif, k=150)
fs.fit(features_200_X,y)
cols = fs.get_support(indices=True)
features_150_X = features_200_X.iloc[:,cols]
```

```python
# Create 'models' dictionary and store all machine learning models in the dictionary
models = {}

# Logistic Regression
from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

# Support Vector Machines
from sklearn.svm import LinearSVC
models['Support Vector Machines'] = LinearSVC()

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
models['K-Nearest Neighbor'] = KNeighborsClassifier()

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

# Import 'cross_val_score' from 'sklearn' library
from sklearn.model_selection import cross_val_score

# Create dictionaries to store perfomance measures
accuracy, recall = {}, {}
std_accuracy, std_recall = {}, {}
full_accuracy, full_recall = {}, {}

# Iterating through every machine learning model in 'models' dictionary
for x in ['250', '200', '150', '100', '50', '25']:

    for key in models.keys():
          
        fs = 'features_' + x + '_X'
        
        # Fit the classifier
        accuracy[key] = cross_val_score(models[key], locals()[fs], y, cv=5, scoring='accuracy').mean()*100
        recall[key] = cross_val_score(models[key], locals()[fs], y, cv=5, scoring='recall').mean()*100

        std_accuracy[key] = cross_val_score(models[key], locals()[fs], y, cv=5, scoring='accuracy').std()*100
        std_recall[key] = cross_val_score(models[key], locals()[fs], y, cv=5, scoring='recall').std()*100

        full_accuracy[key] = str(np.around(accuracy[key],2)) + ' ± ' + str(np.around(std_accuracy[key],2))
        full_recall[key] = str(np.around(recall[key],2)) + ' ± ' + str(np.around(std_recall[key],2))
    
    fs_performance = 'performance_' + x + '_features'
    # Generate performace matrix for every machine learning model
    locals()[fs_performance] = pd.DataFrame(index=models.keys(), columns=['Accuracy','Recall'])
    locals()[fs_performance]['Accuracy'] = accuracy.values()
    locals()[fs_performance]['Recall'] = recall.values()
    
    fs_std = 'std_' + x + '_features'
    # Generate standard deviation matrix for every machine learning model
    locals()[fs_std] = pd.DataFrame(index=models.keys(), columns=['Accuracy','Recall'])
    locals()[fs_std]['Accuracy'] = std_accuracy.values()
    locals()[fs_std]['Recall'] = std_recall.values()
    
    fs_full = 'full_' + x + '_features'
    # Generate standard deviation matrix for every machine learning model
    locals()[fs_full] = pd.DataFrame(index=models.keys(), columns=['Accuracy','Recall'])
    locals()[fs_full]['Accuracy'] = full_accuracy.values()
    locals()[fs_full]['Recall'] = full_recall.values()
```
